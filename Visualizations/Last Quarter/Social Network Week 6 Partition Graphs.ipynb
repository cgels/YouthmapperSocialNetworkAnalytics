```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
import math
import deepgraph as dg
from mpl_toolkits.basemap import Basemap

%matplotlib inline

plt.style.use("ggplot")
```

```python
count = pd.read_csv("SELECT_Z_auth_A__Z_auth_B__sum_1__as_cnt.csv")
```

```python
users = pd.read_csv("SELECT_u_name__ST_X_chap_location__geome.csv", 
                    header=None, 
                    names = ["auth_id", "long", "lat"])
```

```python
# Longitude and latitude correspond to the Xs and Ys 
user_counts = pd.merge(count, users, 
         left_on = "auth_a", right_on = "auth_id")\
.drop("auth_id", axis = 1)
```

```python
graph = dg.DeepGraph(user_counts)
```

```python
graph.v.sort_values("cnt", inplace = True)
```

```python
graph.create_edges_ft(ft_feature=("cnt", 1000))
```

```python
graph.v.head()
```

```python
lv, le = graph.partition_graph("auth_a", 
                      relation_funcs = {"ft_r":["sum", "mean", "std"]})
```

```python
lg = dg.DeepGraph(lv, le)
```
